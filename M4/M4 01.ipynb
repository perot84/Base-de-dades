Una variable categòrica (Districte o Barri)

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
p_lloguer=pd.read_csv ("preu lloguer per trimestre i barri.csv")
districte=p_lloguer.groupby ("Nom_Districte")['Any'].count ()
districte.plot (kind='bar')
plt.title ('Immobles segons districte')
plt.xlabel ('Districte')
plt.savefig("Immobles segons districte.png", bbox_inches='tight')
plt.show()
```

Una variable numèrica (Preu)

```python
p_lloguer['Preu'] = pd.to_numeric(p_lloguer['Preu'],errors = 'coerce')
p_lloguer_mensual = p_lloguer['Lloguer_mitja']=='Lloguer mitjà mensual (Euros/mes)'
p_lloguer ['Preu_mensual']= p_lloguer_mensual
preus_mensuals = p_lloguer['Preu_mensual']==True
preus_mensuals_filtrat = p_lloguer [preus_mensuals]
grafic=sns.histplot(data=preus_mensuals_filtrat, x="Preu", bins =10)
grafic.set(title ='Preu per mes')
plt.savefig("Preu per mes.png", bbox_inches='tight')

plt.show()
```

```python
preus_m2 = p_lloguer ['Preu_mensual']==False
preus_m2_filtrat = p_lloguer [preus_m2] 
grafic2=sns.histplot(data=preus_m2_filtrat, x="Preu", bins =10)
grafic2.set(title ='Preu per m2')
plt.savefig("Preu per m2.png", bbox_inches='tight')
plt.show()
```

Una variable numèrica i una categòrica (Districte i Preu)

```python
grafic3 = preus_mensuals_filtrat.groupby ("Nom_Districte")['Preu'].mean()
grafic3.plot (kind='bar')
plt.title ('Preu mitjà mensual')
plt.xlabel ('Districte')
plt.savefig("Preu mitjà mensual.png", bbox_inches='tight')
plt.show()
```

```python
grafic4 = preus_m2_filtrat.groupby ("Nom_Districte")['Preu'].mean()
grafic4.plot (kind='bar')
plt.title ('Preu mitjà m2')
plt.xlabel ('Districte')
plt.savefig("Preu mitjà m2.png", bbox_inches='tight')
plt.show()
```

```python
sant_marti_barris_m2 = Sant_Martí_filtrat.groupby ("Nom_Barri")['Preu'].mean()
sant_marti_barris_m2.plot (kind='bar')
plt.title ('Preu mitjà m2 al districte de Sant Martí')
plt.xlabel ('Barris')
plt.savefig("Preu mitjà m2 al districte de Sant Martí.png", bbox_inches='tight')
plt.show()
```

Seria interessant fer tots els gràfics per cada districte<br>
També cercar el barri més car i el barri més barat<br>
No hi ha grans diferències segons el semestre per la qual cosa crec que no és important de cara a fer una visualització on hi hagi aquesta variable<br>
També es podria cercar la diferència entre el barri més car i el més barat de cada districte i fer-ne una comparativa per tal de veure a dins de quins districtes hi ha més desigualtat i comparar-lo amb la mitjana dels districtes.

Diferència entre el barri més car i el més barat de per districte (he utilitzat Sant Martí, Horta-Guinardó i Ciutat Vella) i diferència amb la mitjana

```python
preu_max_sant_marti=sant_marti_barris_m2.max()
preu_min_sant_marti=sant_marti_barris_m2.min()
diferència_preus_sant_marti = preu_max_sant_marti - preu_min_sant_marti
```

```python
horta_guinardo_barris_m2 = Horta_Guinardó_filtrat.groupby ("Nom_Barri")['Preu'].mean()
preu_max_horta_guinardo= horta_guinardo_barris_m2.max()
preu_min_horta_guinardo = horta_guinardo_barris_m2.min()
diferència_preus_horta_guinardo = preu_max_horta_guinardo-preu_min_horta_guinardo
```

```python
ciutat_vella_barris_m2 = CV_filtrat.groupby ("Nom_Barri")['Preu'].mean()
preu_max_ciutat_vella= ciutat_vella_barris_m2.max()
preu_min_ciutat_vella = ciutat_vella_barris_m2.min()
diferència_preus_ciutat_vella = preu_max_ciutat_vella-preu_min_ciutat_vella
```

```python
mitjana_dif_preus = (diferència_preus_ciutat_vella + diferència_preus_horta_guinardo + diferència_preus_sant_marti) / 3
```

```python
comparativa_dif_preus_districte = {'Districte': ['Sant martí', 'Horta-Guinardó', 'Ciutat Vella'],
                    'Diferència_Preus': [diferència_preus_sant_marti, diferència_preus_horta_guinardo, diferència_preus_ciutat_vella],
                                  'Mitjana':[mitjana_dif_preus,mitjana_dif_preus,mitjana_dif_preus]}

comparativa_dif_preus_districte = pd.DataFrame(comparativa_dif_preus_districte)
comparativa_dif_preus_districte.set_index('Districte', inplace = True)
```

```python
comparativa_dif_preus_districte.plot (kind='bar', legend=None)
plt.title ('Diferència preus per districte')
plt.xlabel ('Districte')
plt.savefig("Diferència preus per districte.png", bbox_inches='tight')
plt.show()
```

Les dues informacions que crec més rellevants és veure la comparativa entre districtes envers el preu per m2 i la diferència interna de cada districte per tal de detectar on hi ha majors desigualtats
